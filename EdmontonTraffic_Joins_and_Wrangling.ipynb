<h1>Edmonton Photo Radar and Red Light Camera Ticketing</h1>
Merging three Edmonton Open Data datasets from I explore the photo radar ticketing in Edmonton. W
Datasets include: red light cameras study, DriveSafe photo radar trucks trucks and Average Annual Weekday Traffic Volumes. In total, 6 million rows of data merged on longitude and latitude, bringing in approximately 113,000 rows with a match between sets. There is still plenty of data to work with for an anlysis. 
<h3> Assumptions </h3>
</u>
<li> All photo radar trucks operate on the same rules. No operator is making their own decisions on giving people tickets, the machine is treating all vehicles the same.
<li> All red light cameras are programmed the same. If one decides to not give a ticket when going 1-5 km over the speed limit then none of them do. 




```python
df1=pd.read_csv("data/ETickets_2019.csv")
df1.head()
```

```python
df1=df1.drop(columns={"Road Name"})
df1["Red_Truck"]=0
df4= df1.pop("Deployment Hours")
df1["Deployment Hours"]= df4
```

```python
df3=pd.read_csv("data/Red_Light Cams_Intersection_Safety_Device_Tickets_Issued__by_Month_.csv")
```

```python
df5= df3.pop("Total Red Light Tickets")
df3=df3.drop(columns={"New Georeferenced Column"})
df3["Red_Truck"]= 1 
df3["Deployment Hours"]=24
```

```python
df3['Total Red Light Tickets']= df5
df1["Total Red Light Tickets"] = 0
```

```python
df3.head()
```

```python
df1.head()
```

```python
dfconcat=pd.concat([df1, df3])
```

```python
dfconcat.sample(frac=.1)
```

```python
dfconcat.info()
```

```python
df2=pd.read_csv("data/Average_Annual_Weekday_Traffic_Volumes__2009-2015_.csv")
```

```python
df2=df2.drop(columns={"Site Number", "Site Name","Location", "Geometry Point"})
#dfconcat_samp=dfconcat.sample(frac=.1)
df2.rename(columns = {'Latitude':'Latitude1', 'Longitude':'Longitude1'}, inplace = True)
```

```python
df=(dfconcat.assign(dummy=1)
    .merge(df2.assign(dummy=1),on='dummy')
    .query('abs(Latitude-Latitude1)<=0.01 and abs(Longitude-Longitude1)<=0.01')
    .drop('dummy', axis=1)
)
```

```python
df=df.drop(columns={"Latitude1", "Longitude1"})
```

```python
df.head()
```

```python
df['Month'] = pd.DatetimeIndex(df['Month End Date']).month
df['Year'] = pd.DatetimeIndex(df["Month End Date"]).year
#df = df.drop(columns='Month End Date')
df.head()
```

```python
df.isnull().sum()
```

```python
df=df.drop(columns={'Row ID'})
df = df.dropna()
df.isnull().sum()
```
